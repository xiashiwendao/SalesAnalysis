```python

y = groupall.values[:, 2:]
width=16
height = 2
# 正常
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, y[i])
# 取对数
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, np.log(y[i]))

groupall_diff = groupall.diff(axis=1)
groupall_diff.fillna(0, inplace=True)
groupall_diff
# 差分正常
y = groupall_diff.values[:, 2:]
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, y[i])
# 差分对数
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, np.log(y[i]))
    
groupall_diff2 = groupall_diff.diff(axis=1)
groupall_diff2.fillna(0, inplace=True)
# 二次差分正常
y = groupall_diff2.values[:, 2:]
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, y[i])
# 二次差分对数
plt.figure(figsize=(width, height))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.plot(x, np.log(y[i]))
```

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(n_iter_no_change=5, penalty=None, eta0=0.1)


y = groupall.values[:, 2:]
month_count = np.shape(y)[1]
x = range(month_count)
x_array = np.array(x)
x_array_reshape = x_array.reshape(-1, 1)
y_array_reshape = y[0].reshape(-1,1)
sgd_reg.fit(x_array_reshape, y_array_reshape)
predict_months = [[month_count], [month_count + 1], [month_count + 2]]
predict_values = sgd_reg.predict(predict_months)
print(predict_values)
```

```python
groupall = data.groupby(["Region"])
#groupall = data.groupby(["Region", "Sub Region"])
# groupall = data.groupby(["Region", "Sub Region", "MaterialName (EN)"])
groupall = groupall.sum()
#print(groupall.info())

from sklearn.preprocessing import PolynomialFeatures
test_data_count = 2 # 数据样本中，测试样本个数(比如3月份数据做为测试，count=1，拿1，2，3个月数据做测试，count=3)
y_salesForManth = groupall.values[:, 2:] # y是201802~201903的销售值
group_count = np.shape(y_salesForManth)[0] # 分组个数（决定了要循环遍历次数）
#group_count = 1
print("group_count: ", group_count)
x_month_range = range(np.shape(y)[1] -1 * 3) # x是月份值（x轴）
total_score = 0
for i in range(group_count):
    y_train = y_salesForManth[i, :(-1 * test_count)]
    y_test = y_salesForManth[i, (-1 * test_count):]
    y_test[y_test == 0] = 0.1
    plt.plot(x_month_range, y_train, "b.")
    x2 = np.array(x_month_range).reshape(-1,1)
    y2 = np.array(y_train).reshape(-1,1)
    plt.plot(x2, y2, "r.")
    sgd_reg2 = SGDRegressor(random_state=42)
    sgd_reg2.fit(x2, y2.ravel())

    y_predict_fit = sgd_reg2.predict(x2)
    plt.plot(x2, y_predict_fit, "y-")

    y_predict_future = sgd_reg2.predict([[12], [13], [14]])
    acc = 1 - (np.abs(y_predict_future - y_test)/y_test)
    total_score += acc
    #print("accuraty rate: ", acc)
    
avg_score = total_score/group_count
print("avg_score", avg_score)
```
